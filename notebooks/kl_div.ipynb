```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x : np.log2(x)
X = np.arange(0.1, 20, 0.2)
Y = [f(x) for x in X]

plt.xticks([0,1,5,10,15,20])
plt.axhline(0, color='grey', linewidth=0.8)
plt.axvline(0, color='grey', linewidth=0.8)
plt.axvline(1, color='yellow', linewidth=0.8)
plt.plot(X, Y, color='blueviolet')
```

```python
f = lambda x : -np.log2(x)
X = np.arange(0.1, 1, 0.02)
Y = [f(x) for x in X]

plt.xticks([0,1,5,10,15,20])
plt.axhline(0, color='grey', linewidth=0.8)
plt.axvline(0, color='grey', linewidth=0.8)
plt.axvline(1, color='yellow', linewidth=0.8)
plt.plot(X, Y, color='blueviolet')
```

```python
f = lambda x: -np.log10(x)
X = np.arange(0.1, 1, 0.02)
Y = [f(x) for x in X]

plt.xlabel('p(x)')
plt.ylabel('h(x)')
plt.plot(X, Y, color='blueviolet')
```

```python
X = np.arange(1, 20)
Y = [3]*len(X)

plt.plot(X, Y, color='blueviolet')
```

```python
gaussian = lambda x: 1/np.sqrt(2*np.pi) * np.exp(-x**2/2)
X = np.arange(-10, 10, 0.01)
Y = [gaussian(x) for x in X]

plt.plot(X, Y, color='blueviolet')
```
